Mostly performed SMOTE and dataset balancing in this file

```python
import numpy as np
import pandas as pd
```

```python
x_train = np.load('../Data/saved_numpy_arrays/X_psd_epilepsy_corpus.npy', allow_pickle=True)
y_train = np.load('../Data/saved_numpy_arrays/y_epilepsy_corpus.npy', allow_pickle=True)
```

```python
print(x_train.shape)
```

```python
di = dict()
for i in range(len(y_train)):
    if y_train[i] not in di.keys():
        di[y_train[i]] = []
        temp = [[0]*19]*6
        di[y_train[i]]= temp
```

```python
for i in range(len(y_train)):
    for j in range(19):
        for k in range(6):
            if i >= y_train.size: break
            if j >= 19: break
            if k >= 6: break
            if type(x_train[i][j*6+ k]) is tuple or type(x_train[i][j*6+ k]) is list:
                di[y_train[i]][k][j] += x_train[i][j*6+ k][0]
            else:
                di[y_train[i]][k][j] += x_train[i][j*6+ k]
```

```python
for i in di.keys():
    di[i] = np.array(di[i])
    di[i] /= x_train.size
```
